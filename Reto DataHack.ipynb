```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("yasinnaal/bikes-sales-sample-data")

print("Path to dataset files:", path)
```

```python
import os

# Listar los archivos disponibles en el directorio del dataset
archivos = os.listdir(path)
print("Archivos en el dataset:", archivos)
```

```python
import pandas as pd

# Mostrar las columnas de cada archivo CSV en el dataset
for archivo in archivos:
    if archivo.endswith('.csv'):
        try:
            df_temp = pd.read_csv(os.path.join(path, archivo), nrows=0)
        except UnicodeDecodeError:
            df_temp = pd.read_csv(os.path.join(path, archivo), nrows=0, encoding='latin1')
        print(f"Columnas en {archivo}: {list(df_temp.columns)}")
```

```python
# Visualizar cada archivo CSV como una tabla
for archivo_csv in archivos:
    if archivo_csv.endswith('.csv'):
        print(f"Tabla: {archivo_csv}")
        try:
            df = pd.read_csv(os.path.join(path, archivo_csv))
        except UnicodeDecodeError:
            df = pd.read_csv(os.path.join(path, archivo_csv), encoding='latin1')
        display(df.head())
        print("\n" + "-"*60 + "\n")
```

```python
# Limpiar y organizar cada archivo CSV en el dataset
dataframes = {}

for archivo_csv in archivos:
    if archivo_csv.endswith('.csv'):
        try:
            df_temp = pd.read_csv(os.path.join(path, archivo_csv))
        except UnicodeDecodeError:
            df_temp = pd.read_csv(os.path.join(path, archivo_csv), encoding='latin1')
        
        # Eliminar columnas completamente vacas
        df_temp = df_temp.dropna(axis=1, how='all')
        # Eliminar filas completamente vacas
        df_temp = df_temp.dropna(axis=0, how='all')
        # Opcional: resetear el ndice
        df_temp = df_temp.reset_index(drop=True)
        # Guardar el dataframe limpio en un diccionario
        dataframes[archivo_csv.replace('.csv', '')] = df_temp

# Mostrar un resumen de los dataframes limpios
for nombre, df_limpio in dataframes.items():
    print(f"Tabla: {nombre}")
    display(df_limpio.head())
    print(df_limpio.info())
    print("\n" + "-"*60 + "\n")
```

```python
# Agrupar clientes por ventas totales y ordenarlos de mayor a menor
clientes_df = dataframes['BusinessPartners']
ventas_df = dataframes['SalesOrders']

# Agrupar ventas por PARTNERID y sumar el monto bruto
ventas_por_cliente = ventas_df.groupby('PARTNERID')['GROSSAMOUNT'].sum().reset_index()

# Unir con la tabla de clientes para obtener informacin adicional
clientes_ventas = clientes_df.merge(ventas_por_cliente, left_on='PARTNERID', right_on='PARTNERID', how='left')

# Ordenar de mayor a menor ventas
clientes_ventas = clientes_ventas.sort_values(by='GROSSAMOUNT', ascending=False)

display(clientes_ventas)
```

```python
# Obtener los clientes que nunca han realizado compras
clientes_sin_compras = clientes_df[~clientes_df['PARTNERID'].isin(ventas_por_cliente['PARTNERID'])]

display(clientes_sin_compras)
print(f"Total de clientes sin compras: {len(clientes_sin_compras)}")
```

```python
# Calcular ventas totales por mes
ventas_df['MES'] = ventas_df['CREATEDAT'].astype(str).str[:6]  # YYYYMM
ventas_por_mes = ventas_df.groupby('MES')['GROSSAMOUNT'].sum().reset_index()
ventas_por_mes['MES'] = pd.to_datetime(ventas_por_mes['MES'], format='%Y%m')

display(ventas_por_mes)

# Graficar meses de mayor vs menor rendimiento
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(ventas_por_mes['MES'].dt.strftime('%Y-%m'), ventas_por_mes['GROSSAMOUNT'], color='skyblue')
plt.xticks(rotation=45)
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Mes')
plt.tight_layout()
plt.show()
```

```python
# Graficar el rendimiento general de la tienda a lo largo del tiempo

import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(ventas_por_mes['MES'], ventas_por_mes['GROSSAMOUNT'], marker='o', color='teal')
plt.title('Rendimiento General de la Tienda a lo Largo del Tiempo')
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Identificar los aos con ms y menos ventas totales
ventas_df['AO'] = ventas_df['MES'].astype(str).str[:4].astype(int)
ventas_por_ao = ventas_df.groupby('AO')['GROSSAMOUNT'].sum().reset_index()

anio_mas_ventas = ventas_por_ao.loc[ventas_por_ao['GROSSAMOUNT'].idxmax(), 'AO']
anio_menos_ventas = ventas_por_ao.loc[ventas_por_ao['GROSSAMOUNT'].idxmin(), 'AO']

# Cargar los items de las rdenes de venta
items_df = dataframes['SalesOrderItems']

# Unir ventas con items para obtener productos vendidos por ao
ventas_items = ventas_df[['SALESORDERID', 'AO']].merge(
    items_df[['SALESORDERID', 'PRODUCTID', 'GROSSAMOUNT']],
    on='SALESORDERID',
    how='inner'
)

# Productos ms vendidos en el ao con menos ventas
productos_menos = ventas_items[ventas_items['AO'] == anio_menos_ventas] \
    .groupby('PRODUCTID')['GROSSAMOUNT'].sum().reset_index() \
    .sort_values(by='GROSSAMOUNT', ascending=False)

# Productos ms vendidos en el ao con ms ventas
productos_mas = ventas_items[ventas_items['AO'] == anio_mas_ventas] \
    .groupby('PRODUCTID')['GROSSAMOUNT'].sum().reset_index() \
    .sort_values(by='GROSSAMOUNT', ascending=False)

print(f"Ao con menos ventas: {anio_menos_ventas}")
display(productos_menos.head(10))

print(f"Ao con ms ventas: {anio_mas_ventas}")
display(productos_mas.head(10))

# Tabla resumen de productos ganadores de ambos aos
productos_ganadores = pd.DataFrame({
    'Ao': [anio_menos_ventas, anio_mas_ventas],
    'Producto ms vendido': [
        productos_menos.iloc[0]['PRODUCTID'],
        productos_mas.iloc[0]['PRODUCTID']
    ],
    'Ventas': [
        productos_menos.iloc[0]['GROSSAMOUNT'],
        productos_mas.iloc[0]['GROSSAMOUNT']
    ]
})

display(productos_ganadores)
```

```python
# Ventas totales por categora de producto

# Unir items_df con products_df para obtener la categora de cada producto
products_df = dataframes['Products']
product_categories_df = dataframes['ProductCategories']
product_category_text_df = dataframes['ProductCategoryText']

# Unir para obtener la categora de cada producto
items_with_category = items_df.merge(
    products_df[['PRODUCTID', 'PRODCATEGORYID']],
    on='PRODUCTID',
    how='left'
).merge(
    product_category_text_df[['PRODCATEGORYID', 'SHORT_DESCR']],
    on='PRODCATEGORYID',
    how='left'
)

# Agrupar por categora y sumar ventas
ventas_por_categoria = items_with_category.groupby('SHORT_DESCR')['GROSSAMOUNT'].sum().reset_index().sort_values(by='GROSSAMOUNT', ascending=False)

display(ventas_por_categoria)

# Graficar ventas por categora
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(ventas_por_categoria['SHORT_DESCR'], ventas_por_categoria['GROSSAMOUNT'], color='mediumseagreen')
plt.xlabel('Categora de Producto')
plt.ylabel('Ventas Totales')
plt.title('Ventas por Categora de Producto')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Identificar los tipos de productos con mayor demanda (ms vendidos por cantidad)

# Agrupar items por categora y sumar la cantidad vendida
productos_mas_demanda = items_with_category.groupby('SHORT_DESCR')['QUANTITY'].sum().reset_index()
productos_mas_demanda = productos_mas_demanda.sort_values(by='QUANTITY', ascending=False)

display(productos_mas_demanda)

# Graficar los productos ms demandados por cantidad
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(productos_mas_demanda['SHORT_DESCR'], productos_mas_demanda['QUANTITY'], color='cornflowerblue')
plt.xlabel('Tipo de Producto')
plt.ylabel('Cantidad Vendida')
plt.title('Demanda por Tipo de Producto')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calcular el stock disponible por producto usando la tabla 'Products'
# Se asume que la columna 'PRICE' en products_df es el precio unitario, y 'QUANTITYUNIT' es la unidad de medida.
# Si el stock no est explcito, se puede calcular la suma de QUANTITY vendida por producto y restar de un stock inicial si existiera.
# Aqu solo mostramos la cantidad vendida por producto como referencia de movimiento.

stock_por_producto = items_df.groupby('PRODUCTID')['QUANTITY'].sum().reset_index()

# Merge with products_df to get QUANTITYUNIT
stock_por_producto = stock_por_producto.merge(
    products_df[['PRODUCTID', 'QUANTITYUNIT']],
    on='PRODUCTID',
    how='left'
)

# Merge with ProductTexts to get product-level SHORT_DESCR
product_texts_df = dataframes['ProductTexts']
stock_por_producto = stock_por_producto.merge(
    product_texts_df[['PRODUCTID', 'SHORT_DESCR']],
    on='PRODUCTID',
    how='left'
)

stock_por_producto = stock_por_producto.rename(columns={'QUANTITY': 'CANTIDAD_VENDIDA'})

display(stock_por_producto)
```
